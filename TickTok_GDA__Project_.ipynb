```python
# Calculate median views by author ban status
ban_status_median_views = df.groupby('author_ban_status')['video_view_count'].median().reset_index()

# Create bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=ban_status_median_views, x='author_ban_status', y='video_view_count', hue= 'author_ban_status' )
plt.title('Median Video Views by Author Ban Status')
plt.xlabel('Author Ban Status')
plt.ylabel('Median View Count')
plt.show()
```

```python
# Create a pie graph that depicts the proportions of total views for claim videos and total views for opinion videos.

claim_status_proportions = df.groupby('claim_status')['video_view_count'].sum()

plt.figure(figsize=(8, 5))
plt.pie(claim_status_proportions, labels=claim_status_proportions.index)
plt.title('Proportions of Total Views for Claim and Opinion Videos')
plt.show()
```

```python
# List of count columns to analyze
count_cols = ['video_view_count', 'video_like_count', 'video_share_count',
              'video_download_count', 'video_comment_count']

for column in count_cols:
    # Calculate IQR and Median
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    median = df[column].median()

    # Calculate outlier threshold (median + 1.5 * IQR)
    threshold = median + (1.5 * iqr)

    # Calculate number of videos exceeding the threshold
    outlier_count = (df[column] > threshold).sum()

    print(f"Number of outliers, {column}: {outlier_count}")
```

```python
# Create a scatterplot of `video_view_count` versus `video_like_count` according to 'claim_status'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='video_view_count', y='video_like_count', hue='claim_status', s=10, alpha=0.3)
plt.title('Video View Count vs Video Like Count by Claim Status')
plt.xlabel('Video View Count')
plt.ylabel('Video Like Count')
plt.legend(title='Claim Status')
plt.show()
```

```python
# Create a scatterplot of `video_view_count` versus `video_like_count` for opinions only
opinion_df = df[df['claim_status'] == 'opinion']
plt.figure(figsize=(10, 6))
sns.scatterplot(data=opinion_df, x='video_view_count', y='video_like_count', s=10, alpha=0.3, color='orange')
plt.title('Video View Count vs Video Like Count (Opinions Only)')
plt.xlabel('Video View Count')
plt.ylabel('Video Like Count')
plt.show()
```
